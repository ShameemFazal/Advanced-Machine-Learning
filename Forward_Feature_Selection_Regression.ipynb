```python
#importing the libraries
import pandas as pd
```

```python
# Loading the dataset
dataset1=pd.read_csv("prep.csv",index_col=None)
df2=dataset1
df2 = pd.get_dummies(df2, drop_first=True)
print(df2.shape)
```

```python
# checking the  missing values in the dataset
df2.isnull().sum()
```

```python
# Defining Target and Independent Variables
indep_X=df2.drop('classification_yes',axis=1)
dep_Y=df2['classification_yes']
indep_X
```

```python
# importing the models

from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.linear_model import LinearRegression
```

```python
# calling the linear regression model

lreg = LinearRegression()
sfs1 = sfs(lreg, k_features=4, forward=True, verbose=2, scoring='neg_mean_squared_error')
```

```python
sfs1 = sfs1.fit(indep_X,dep_Y)
```

```python
feat_names = list(sfs1.k_feature_names_)
print(feat_names)
```

```python
# creating a new dataframe using the above variables and adding the target variable
new_data = df2[feat_names]
new_data['classification_yes'] = df2['classification_yes']

# first five rows of the new data
new_data.head()
```
